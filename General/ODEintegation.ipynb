````python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
````

## ODE integrators ##

Many problems in physics reduce to solving ordinary differential equations (see, e.g., all of mechanics!) and frequently these are initial value problems.  While there is an enormous and sophisticated theory and practice for ODE integration, for much of the time you can in fact get by with just a couple of stand-bys.

For definiteness let us imagine we are interested in the solution of the ODE
$$
\frac{d\vec{y}}{dt} = f(t,\vec{y})
$$
given (well posed) initial values $\vec{y}(t_0)$.  We imagine taking steps, of size $h$, in $t$ and updating $\vec{y}$ through a sequence of values $\vec{y}_0$, $\vec{y}_1$, ..., etc.

First off, if you know something about the natural timescale of your ODE and you know it's not going to change dramatically then it's hard to do better than [fourth order Runge-Kutta method](https://en.wikipedia.org/wiki/Runge%E2%80%93Kutta_methods), colloquially known as [RK4](https://en.wikipedia.org/wiki/List_of_Runge%E2%80%93Kutta_methods).  This should be in everyone's toolkit.

````python
def deriv(t,y):
    """Returns dy/dt for a cosine/sine system as an example."""
    dydt    = np.zeros_like(y)
    dydt[0] = -y[1]
    dydt[1] =  y[0]
    return(dydt)
    #

def rk4(t,y,h):
    """Makes a single RK4 step."""
    k1 = h * deriv(t      ,y)
    k2 = h * deriv(t+0.5*h,y+k1/2)
    k3 = h * deriv(t+0.5*h,y+k2/2)
    k4 = h * deriv(t+    h,y+k3)
    return( (t+h,y+(k1+2*k2+2*k3+k4)/6.0) )
````

Note that the limit that $f(t,\vec{y})$ is independent of $\vec{y}$ this becomes Simpson's rule for integration which may make RK4 easier to remember.

````python
# Let's test it ...
tarr= []
yarr= []
#
h   = 0.2
t,y = 0.0,np.array([1.0,0.0])
for i in range(100):
    t,y = rk4(t,y,h)
    tarr.append(t)
    yarr.append(y[0])
tarr = np.array(tarr)
yarr = np.array(yarr)
#
fig,ax = plt.subplots(1,1,figsize=(6,4))
ax.plot(tarr,yarr,label='RK4')
ax.plot(tarr,np.cos(tarr),label='Exact')
ax.legend()
ax.set_xlabel('$t$')
ax.set_ylabel('$y(t)$')
#
ap = yarr[-1]
ex = np.cos(tarr[-1])
print("Error at end is {:f}%".format(100*(ap-ex)/ex))
````

This is actually pretty good performance, and if you know something about the period then you can dial the accuracy you want by just setting $h$ appropriately.

### Adaptive steps ###

For a more general case, one wants to use some form of adaptive stepsize control.  Unless you have an external way of guessing a good stepsize (and sometimes you do!) then the standard methods are

1) try a step with size $h$ and $h/2$, compare the results and decide whether to keep it or not OR

2) embed a lower order integrator inside your higher order one and compare the two at the end of the step.

A very common example of (2) is a combined 4th order and 5th order Runge-Kutta.  A cute example which only involves 3 function evaluations per step is a combination of 2nd and 3rd order integrators known as [Bogacki Shampine](https://en.wikipedia.org/wiki/Bogacki%E2%80%93Shampine_method).

````python
def ode23(t,y,h,k4):
    """A single step of the Bogacki Shampine method -- note the last
    step of this method computes a "k4" which equals the "k1" of the
    next step, so we take it in and return it.  The next step would
    be determined by the value of the error you get back (i.e. the
    difference between "y" and "z")."""
    y0 = y
    k1 = k4
    k2 = h * deriv(t+0.50*h,y+0.50*k1)
    k3 = h * deriv(t+0.75*h,y+0.75*k2)
    y  = y + (2*k1+3*k2+4*k3)/9.0
    k4 = h * deriv(t+h,y)
    z  = y0+ (7*k1+6*k2+8*k3+3*k4)/24.0
    return( (t+h,y,z,k4) )
````

````python
# Let's test it ...
tarr= []
yarr= []
earr= []
#
h   = 0.2
t,y = 0.0,np.array([1.0,0.0])
k1  = h*deriv(t,y)
for i in range(100):
    t,y,z,k1 = ode23(t,y,h,k1)
    tarr.append(t)
    yarr.append(y[0])
    earr.append(y[0]-z[0])
tarr = np.array(tarr)
yarr = np.array(yarr)
earr = np.array(earr)
#
fig,ax = plt.subplots(1,1,figsize=(6,4))
ax.plot(tarr,yarr,label='BS')
ax.plot(tarr,np.cos(tarr),label='Exact')
ax.plot(tarr,1e3*earr,label=r'$1000\times$ Err')
ax.legend()
ax.set_xlabel('$t$')
ax.set_ylabel('$y(t)$')
#
ap = yarr[-1]
ex = np.cos(tarr[-1])
print("Error at end is {:f}%".format(100*(ap-ex)/ex))
````

One advantage of this low-order method is that if the step size is changing rapidly and unexpectedly you are only making 3 function evaluations that you throw away.  The disadvantage if that if the ODE is mostly well behaved then you're not getting the full power of a high-order integrator.

Here's an example of how one implements the adaptive stepsize ... this particular system is so simple it really only adjusts once, but this gives the idea:

````python
# Now do adaptive step sizes using a wrapper.
abstol = 1e-5
reltol = 1e-7
#
tarr,yarr = [],[]
#
h   = 0.01 # A first guess to get us going ...
t,y = 0.0,np.array([1.0,0.0])
k1  = h*deriv(t,y)
while t<2.0:
    tnew,ynew,znew,k1new = ode23(t,y,h,k1)
    tol = abstol + np.max([np.abs(ynew),np.abs(znew)]) * reltol
    err = np.max(np.abs(ynew-znew)) + 1e-30
    print("{:.3f} {:12.4e} {:12.4e} {:12.4e}".format(t,h,err,tol))
    if err<tol:
        t,y,z,k1 = tnew,ynew,znew,k1new
        tarr.append(t)
        yarr.append(y[0])
        if err<0.2*tol:
            # If we're doing too well, allow step to grow.
            fac = np.min([(tol/err)**(0.3333),5.0])
            print("-- accepting step, changing h with fac={:f}".format(fac))
            h  *= fac
            k1 *= fac
    else:
        fac = 0.9*(tol/err)**(0.3333)
        print("-- redoing step with fac={:f}".format(fac))
        h  *= fac
        k1 *= fac
tarr = np.array(tarr)
yarr = np.array(yarr)
#
ap = yarr[-1]
ex = np.cos(tarr[-1])
print("Took {:d} steps.".format(tarr.size))
print("Error at end is {:f}%".format(100*(ap-ex)/ex))
````

### Very low order integrators ###

The above uses a combination of a 2nd and a 3rd order integrator.  Sometimes it is better to use a low order ODE integrator and just hammer the number of steps, rather than a higher order integrator that requires fewer function calls.  The lowest order method is Euler's method, which is normally introduced primarily to show how bad it is!  Going up one order, there are two 2nd order methods worth knowing:

The most "obvious" second order method (related to the "leap frog" method for symplectic systems -- see below) is the [midpoint method](https://en.wikipedia.org/wiki/Midpoint_method).  Here we use Euler's method to get an estimate for $\vec{y}$ at the midpoint of the interval, then evaluate $f$ there and use that value to take the step.
A lesser known method, but one which feels very "obvious" in hindsight, is [Heun's method](https://en.wikipedia.org/wiki/Heun's_method#Runge.E2.80.93Kutta_method).  In this method we make an estimate of the slope at the beginning and end of the interval, average them and then take a step using that slope.  See the notebook ```LowOrderODEint``` for more details.

## Symplectic integrators ##

What if the kind of systems we want to evolve come from a Hamiltonian, i.e. they are the equations of motion for some system?  The above integrators would be a reasonable choice, but in fact one can do much, much better using the particular behavior of Hamiltonian flows.

One of the issues with the above integrators is that they don't know about the special structure of a Hamiltonian evolution -- the fact that the equations for $\dot{x}$ and $\dot{p}$ are so similar and that very often we have an exactly conserved quantity: $H$.  What you find is that you can get slow drifts of the energy over time, and orbits which should actually close on themselves can start to drift very far from their proper place.

This leads us to consider symplectic integrators.

Suppose we have a single particle moving under a Hamiltonian $H(q,p)=p^2/(2m) + V(q)$ with
$$
  \dot{p} = -\frac{dH}{dq} = -\frac{dV}{dq}
  \quad , \quad
  \dot{q} = \frac{dH}{dp} = \frac{p}{m}
$$
How should we evolve this system?

### Leapfrog ###

The best low order symplectic integrator for such a system is the [leapfrog scheme](https://en.wikipedia.org/wiki/Leapfrog_integration), which interleaves "drifts" of particles with "kicks" due to the forces:

1) $v_{n+1/2} = v_{n} + (h/2) a_n$

2) $q_{n+1} = q_n + h v_{n+1/2}$

3) $v_{n+1} = v_{n+1/2} + (h/2) a_{n+1}$

where $a_n$ is the acceleration at time $n$.  Note that for constant steps you can combine the first (1) and last (3) step into a single "kick" for time $h$.  This algorithm has two advantages: (a) it is explicitly time reversible and (b) it conserves an energy which is close to the physical energy that we want to conserve.

### Orbits near a Schwarzchild black hole ###

As a slightly non-trivial and physically interesting example let us consider orbits of a test particle near a Schwarzchild black hole.  Since the Schwarzchild geometry is spherically symmetric we can choose spherical coordinates and work in the plane $\theta=\pi/2$.  The effective potential for the radial motion then becomes
$$
  \dot{r}^2 + V_{\rm eff}(r) = k^2
  \quad , \quad
  V_{\rm eff}(r) = \left(1+\frac{\ell^2}{r^2}\right)\left(1-\frac{r_s}{r}\right)
$$
where $r_s$ is the Schwarzchild radius, $k$ is a constant related to energy and $\ell$ is a constant related to angular momentum ($\ell=r^2\dot{\phi}$).  Note that overdot here represents a derivative with respect to the speed of light times the (proper) time, so $\dot{r}$ is dimensionless.

To integrate these equations let us work in units of $r_s$ and consider the system
\begin{eqnarray}
  \dot{r}    &=& p \\
  \dot{p}    &=& -\frac{dV}{dr} = -\frac{1}{r^2} + \frac{\ell^2}{r^4}(2r-3) \\
  \dot{\phi} &=& \frac{\ell}{r^2} \\
  \dot{\ell} &=& 0
\end{eqnarray}
where I have written the $\phi$ equation and the angular momentum out explicitly but we likely wouldn't actually evolve $\dot{\ell}=0$ explicitly.

````python
# Define the constants for our orbit.
ell    = 2.0
r_init = 10.
tmax   = 500.0
````

````python
def deriv(t,y):
    """Returns dy/dt for a Schwarzchild system as an example."""
    dydt    = np.zeros_like(y)
    rr      = y[0]
    oner2   = 1.0/rr**2
    dydt[0] = y[1]
    dydt[1] = -oner2 + ell**2*oner2*oner2*(2*rr-3)
    dydt[2] = ell*oner2
    return(dydt)
    #
````

````python
# Set the initial conditions at the turning point,
# which we call phi=0:
y   = np.array([r_init,0.0,0])
tarr= np.array([0])
earr= np.array([(1.0+ell**2/y[0]**2)*(1-1/y[0])])
yarr= np.zeros((1,3))
yarr[0,:] = y
#
# First integrate this using a "normal" integrator.
# Try RK4 with constant step sizes.
#
t,dt = tarr[0],5.0
while t<tmax:
    t,y = rk4(t,y,dt)
    Veff = (1.0+ell**2/y[0]**2)*(1-1/y[0])
    enrgy= 0.5*y[1]**2+Veff
    tarr = np.append(tarr,t)
    earr = np.append(earr,enrgy)
    yarr = np.append(yarr,y[None,:],axis=0)
# How many steps did we take?
print("Took ",tarr.size," steps.")
print("Final position at t={:f} is (r,phi)={:f},{:f}".\
     format(t,yarr[-1,0],yarr[-1,2]))
````

````python
# Plot the orbit and energy.
xx = yarr[:,0]*np.cos(yarr[:,2])
yy = yarr[:,0]*np.sin(yarr[:,2])
#
fig,ax = plt.subplots(1,2,figsize=(12,4))
ax[0].plot(xx,yy)
ax[0].set_aspect('equal')
ax[0].set_xlabel(r'$x$')
ax[0].set_ylabel(r'$y$')
#
ax[1].plot(tarr,earr-earr[0])
ax[1].set_xlabel(r'$c\tau$')
ax[1].set_ylabel(r'$\Delta$ Energy')
````

__Note__ the "perhelion advance" where the orbit doesn't close on itself (here in a somewhat extreme example)!

It turns out the final position isn't _too_ bad after several orbits ... but take a look at what is happening to the energy.  This should be a conserved quantity but is steadily drifting lower and lower.  In fact it jumps every time we get to an inner turning point where the potential gets "interesting".

Ok, that's for fixed time steps and we could obviously do a bit better by making the time steps smaller.  What happens if we get more sophisticated and try adapative step size control?  We'll use our ODE23 example above:

````python
# Set the initial conditions at the turning point,
# which we call phi=0:
y   = np.array([r_init,0.0,0])
tarr= np.array([0])
earr= np.array([(1.0+ell**2/y[0]**2)*(1-1/y[0])])
yarr= np.zeros((1,3))
yarr[0,:] = y
#
# Try with adaptive stepsize control.  We'll
# use our ode23 example above.
#
t,dt = tarr[0],0.1
k1   = dt*deriv(t,y)
while t<tmax:
    tnew,ynew,znew,k1new = ode23(t,y,dt,k1)
    tol = abstol + np.max([np.abs(ynew),np.abs(znew)]) * reltol
    err = np.max(np.abs(ynew-znew)) + 1e-30
    if err<tol:
        t,y,z,k1 = tnew,ynew,znew,k1new
        #
        Veff = (1.0+ell**2/y[0]**2)*(1-1/y[0])
        enrgy= 0.5*y[1]**2+Veff
        tarr = np.append(tarr,t)
        earr = np.append(earr,enrgy)
        yarr = np.append(yarr,y[None,:],axis=0)
        if err<0.2*tol:
            # If we're doing too well, allow step to grow.
            fac = np.min([(tol/err)**(0.3333),5.0])
            dt *= fac
            k1 *= fac
    else:
        fac = 0.9*(tol/err)**(0.3333)
        dt *= fac
        k1 *= fac
# How many steps did we take?
print("Took ",tarr.size," steps.")
print("Final position at t={:f} is (r,phi)={:f},{:f}".\
     format(t,yarr[-1,0],yarr[-1,2]))
````

````python
# Plot the orbit and energy.
xx = yarr[:,0]*np.cos(yarr[:,2])
yy = yarr[:,0]*np.sin(yarr[:,2])
#
fig,ax = plt.subplots(1,2,figsize=(12,4))
ax[0].plot(xx,yy)
ax[0].set_aspect('equal')
ax[0].set_xlabel(r'$x$')
ax[0].set_ylabel(r'$y$')
#
ax[1].plot(tarr,earr-earr[0])
ax[1].set_xlabel(r'$c\tau$')
ax[1].set_ylabel(r'$\Delta$ Energy')
````

The integrator wanted a lot more steps, and that has certainly led to smoother looking orbits!  We still see the perihelion advance nicely with the orbits wrapping around.  But even though the evolution is now very smooth the energy is still dropping in steps every orbit.

So now let's code this up using the leapfrog integrator ... here we'll take large steps and see what happens.

````python
# Set the initial conditions
y   = np.array([r_init,0.0,0])
tarr= np.array([0])
earr= np.array([(1.0+ell**2/y[0]**2)*(1-1/y[0])])
yarr= np.zeros((1,3))
yarr[0,:] = y
#
t,dt = tarr[0],5.0
accn = -1/y[0]**2 + ell**2/y[0]**4*(2*y[0]-3)
while t<tmax:
    y[1]+= 0.5*dt*accn
    #
    y[2] += dt*ell/y[0]**2
    y[0] += dt*y[1]
    #
    rr    = y[0]
    oner2 = 1.0/rr**2
    accn  = -oner2 + ell**2*oner2*oner2*(2*rr-3)
    y[1] += 0.5*dt*accn
    t    += dt
    #
    Veff = (1.0+ell**2/y[0]**2)*(1-1/y[0])
    enrgy= 0.5*y[1]**2+Veff
    #
    tarr = np.append(tarr,t)
    earr = np.append(earr,enrgy)
    yarr = np.append(yarr,y[None,:],axis=0)
    #
# How many steps did we take?
print("Took ",tarr.size," steps.")
print("Final position at t={:f} is (r,phi)={:f},{:f}".\
     format(t,yarr[-1,0],yarr[-1,2]))
````

````python
# Plot the orbit and energy.
xx = yarr[:,0]*np.cos(yarr[:,2])
yy = yarr[:,0]*np.sin(yarr[:,2])
#
fig,ax = plt.subplots(1,2,figsize=(12,4))
ax[0].plot(xx,yy)
ax[0].set_aspect('equal')
ax[0].set_xlabel(r'$x$')
ax[0].set_ylabel(r'$y$')
#
ax[1].plot(tarr,earr-earr[0])
ax[1].set_xlabel(r'$c\tau$')
ax[1].set_ylabel(r'$\Delta$ Energy')
````

This has actually worked pretty well for a 2nd order method (with only one acceleration calculation per step)!  The positions are looking not too bad, but far more interesting is the energy.  Notice that it is fluctuating around, but the secular trend with time has now disappeared!  It turns out that the method is "correctly" integrating a Hamiltonian system (with conserved energy), it's just that for large-ish time steps the Hamiltonian isn't quite the one we wanted to integrate.

To see this, and to see how to generalize the leapfrog method, let's look at how to build such symplectic integrators ...

## Higher order methods - background ##

The theory of symplectic integrators is large and rewarding, but life is short so we'll take the simplest approach here.  We will build around the idea of `operator splitting` (sometimes known as "composition methods"; other techniques we won't touch are Runge-Kutta and "partitioned" Runge-Kutta).

Basically, evolving an ODE from step $n$ to $n+1$ defines a map from $(q^n,p^n)$ to $(q^{n+1},p^{n+1})$.  For Hamiltonian systems the evolution looks like the exponential of some matrix, $\sim H\Delta t$.  If our Hamiltonian is separable, $H=H_1+H_2$, we can approximate this as integration under $H_1$ followed by integration under $H_2$.  Or we could try to be more symmetric and do half a step under $H_1$, a full step under $H_2$ followed by another half step under $H_1$.

To connect this idea to the above, what if $H_1=p^2/(2m)$ and $H_2=V(q)$?  Evolution under $H_1$ leaves $p$ unchanged, updating $q$, evolution under $H_2$ leaves $q$ unchanged, updating $p$, and we get the leapfrog evolution above.

Aside: The key to the long term stability is that one can use the Baker-Campbell-Haussdorf formula (BCH; probably familiar from quantum mechanics class) to rewrite the product of exponentials as a single exponential:
$$
  e^{tA} e^{tB} = e^{t(A+B) + (1/2)t^2[A,B] + \cdots} \quad\mathrm{(BCH)}
$$
This means our broken up evolution is like evolution under a single Hamiltonian that is "close to" the Hamiltonian $H_1+H_2$.  Our evolution will be "exact" for a Hamiltonian, just not quite the Hamiltonian that we want.  Our numerical Hamiltonian will get closer and closer to our desired Hamiltonian as $\Delta t\to 0$, so we can neglect the higher order terms.

### Composition methods ###

Let us imagine $H$ is seperable into a kinetic and potential energy piece: $H=T(p)+V(q)$.  For each, we can exactly solve the equations:
$$
  \frac{dq}{dt} =  0 \quad , \quad \frac{dp}{dt} = -\nabla_q V(q)
$$
and
$$
  \frac{dq}{dt} = \nabla_p T(p) \quad , \quad \frac{dp}{dt} = 0
$$
We will denote a single step, by time $t$, of the first system by $\Phi_{t,V}$ and of the second $\Phi_{t,T}$.  We're going to consider compositions like:
$$
  \Phi_t = \Phi_{b_N t,V}\Phi_{a_N t,V} \cdots \Phi_{b_1 t,V}\Phi_{a_0 t,V}  
$$
with $a_i$ and $b_i$ some fractions such that $\sum_i a_i = \sum_i b_i = 1$.  Our leapfrog had $a_0=a_1=1/2$ and $b_1=1$.  We can obtain higher order methods by using the BCH formula and setting $a_i$ and $b_i$ to zero out successively higher order powers of $\Delta t$.

An example of a 4th order method is
$$
  a_0 = a_6 = 0.079203696 , a_1 = a_5 = 0.3531729 , a_2 = a_4 = -0.04206508 , \sum a_i=1
$$
and
$$
  b_1 = b_5 = 0.2095151 , b_2 = b_4 = -0.14385177 , \sum d_i=1
$$

Alternatively we can build up higher order integrators by taking our leapfrog method and doing multiple leapfrog steps one after the other but adjusting the step size each time so that by the end of the chain we have eliminated errors up to some power of $\Delta t$ -- see e.g. [McLachlan (1995)](https://epubs.siam.org/doi/10.1137/0916010) or Yoshida (1993).

One thing to notice about these methods is that they tend to do a "backwards in time" integration for some steps (some of the $a$s and $b$s are negative).  This is a general property of these higher-order schemes.

There are also generalizations to systems including constraints etc., but that starts to get pretty specialized.
